##Eksplorasi Data orders_df


Berikut penjelasan detail terkait kode tersebut:

1. Menghitung Selisih Waktu:

delivery_time = orders_df["delivery_date"] - orders_df["order_date"]

Mengambil kolom delivery_date dan order_date dari DataFrame orders_df.
Mengurangi nilai pada kolom order_date dari nilai pada kolom delivery_date.
Menyimpan hasil pengurangan (selisih waktu) dalam variabel delivery_time.
Operasi ini menghasilkan selisih waktu dalam format Timedelta (representasi perbedaan waktu dalam pandas).
2. Konversi Selisih Waktu ke Detik:

delivery_time = delivery_time.apply(lambda x: x.total_seconds())

Menerapkan fungsi lambda (lambda x: x.total_seconds()) pada setiap elemen dalam variabel delivery_time.
Fungsi lambda memanggil metode total_seconds() pada setiap Timedelta untuk mengkonversinya menjadi jumlah detik.
Hasilnya disimpan kembali dalam variabel delivery_time, kini berisi nilai detik.
3. Mengubah Detik ke Hari dan Membulatkan:

orders_df["delivery_time"] = round(delivery_time/86400)

Membagi nilai detik dalam delivery_time dengan 86400 (jumlah detik dalam sehari).
Membulatkan hasil pembagian menggunakan fungsi round() untuk mendapatkan nilai hari utuh.
Menambahkan hasil pembulatan sebagai kolom baru bernama delivery_time dalam DataFrame orders_df.
Intinya:

```python
delivery_time = orders_df["delivery_date"] - orders_df["order_date"]
delivery_time = delivery_time.apply(lambda x: x.total_seconds())
orders_df["delivery_time"] = round(delivery_time/86400)
```

```python
orders_df.describe(include="all")
```

# Eksplorasi Data orders_df dan customers_df




1. Membuat Daftar Customer ID dari DataFrame Orders:

customer_id_in_orders_df = orders_df.customer_id.tolist()

Mengambil kolom customer_id dari DataFrame orders_df.
Mengubah nilai pada kolom tersebut menjadi sebuah list dan menyimpannya dalam variabel customer_id_in_orders_df. List ini berisi semua customer ID yang pernah melakukan pemesanan.

2. Menambahkan Kolom Status Berdasarkan Aktivitas:

customers_df["status"] = customers_df["customer_id"].apply(lambda x: "Active" if x in customer_id_in_orders_df else "Non Active")

Menerapkan fungsi lambda pada setiap elemen dalam kolom customer_id di DataFrame customers_df.
Fungsi lambda memeriksa apakah customer ID tersebut (x) terdapat dalam list customer_id_in_orders_df.
Jika ada, artinya customer tersebut pernah melakukan pemesanan, maka nilai "Active" dimasukkan ke kolom status.
Jika tidak ada, artinya customer belum pernah melakukan pemesanan, maka nilai "Non Active" dimasukkan ke kolom status.
Hasil pemrosesan fungsi lambda ini disimpan sebagai kolom baru bernama status pada DataFrame customers_df.




```python
customer_id_in_orders_df = orders_df.customer_id.tolist()
customers_df["status"] = customers_df["customer_id"].apply(lambda x: "Active" if x in customer_id_in_orders_df else "Non Active")
customers_df.sample(5)
```

```python
customers_df.groupby(by='status').customer_id.count()
```

Seperti yang terlihat pada pivot table di atas, terdapat cukup banyak pelanggan yang belum pernah melakukan transaksi sebelumnya. Ini tentunya merupakan kabar buruk bagi kita karena hampir 30% pelanggan kita belum pernah melakukan order sebelumnya.

Untuk memperoleh lebih banyak informasi terkait kedua data tersebut, kita perlu menggabungkan keduanya melalui proses join atau merge.

```python
orders_customers_df = pd.merge(
    left= orders_df,
    right = customers_df,
    how='left',
    left_on='customer_id',
    right_on='customer_id'
)

orders_customers_df.head()
```

## Jumlah order berdasarkan kota

```python
orders_customers_df.groupby(by='city').order_id.nunique().sort_values(ascending=False).reset_index().head(10)
```

## Jumlah order berdasarkan state

```python
orders_customers_df.groupby(by="state").order_id.nunique().sort_values(ascending=False)
```

## Jumlah order berdasarkan gender

```python
orders_customers_df.groupby(by='gender').order_id.nunique().sort_values(ascending=False)
```

## Jumlah order berdasarkan kelompok usia


```python
orders_customers_df['age_group'] = orders_customers_df.age.apply(lambda x: 'Youth' if x <= 24 else ('Seniors' if x > 64 else 'Adults'))
orders_customers_df.groupby(by='age_group').order_id.nunique().sort_values(ascending=False)
```

 kita bisa menggunakan pivot table untuk mencari informasi terkait produk berdasarkan tipe dan nama produknya.

```python
product_df.groupby(by='product_name').agg({
    'product_id' : 'nunique',
    'quantity' : 'sum',
    'price' : ['min', 'max']
})
```

Sebagai calon praktisi data yang Andal, tentunya Anda penasaran dengan produk yang paling laris. Nah, untuk menjawab pertanyaan ini, kita perlu menyatukan (merge) tabel product_df dan sales_df dengan kode berikut.



Jika diperhatikan kembali, hasil dari proses merge di atas memiliki perbedaan antara nilai price_per_unit dan price. Hal ini bisa terjadi karena adanya potongan harga, biaya operasional, dan biaya lainnya.(penjelasan dari outpuya)

```python
sales_product_df = pd.merge(
    left=sales_df,
    right=product_df,
    how='left',
    left_on='product_id',
    right_on='product_id'
)

sales_product_df.head()
```

Jika Anda perhatikan pivot table di atas, Trousers merupakan tipe produk yang paling laris. Akan tetapi, jika dilihat berdasarkan revenue yang diterima, Jacket merupakan tipe produk yang paling banyak menyumbang revenue perusahaan.(outputnya)

```python
sales_product_df.groupby(by='product_type').agg({
    'sales_id' : 'nunique',
    'quantity_x' : 'sum',
    'total_price' : 'sum'
})
```

Kita bisa membuat pivot table yang sama untuk melihat informasi penjualan berdasarkan nama produk menggunakan kode di bawah ini


Berdasarkan pivot table tersebut, diketahui bahwa produk Denim merupakan produk yang paling laris dan juga paling banyak menyumbang revenue terhadap perusahaan.(outputnya)  





```python
sales_product_df.groupby(by="product_name").agg({
    "sales_id": "nunique",
    "quantity_x": "sum",
    "total_price": "sum"
}).sort_values(by="total_price", ascending=False)
```
